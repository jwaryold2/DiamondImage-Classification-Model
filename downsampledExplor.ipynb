```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
```

```python
# Data from Diamond_Data
colors_list = list(mcolors.CSS4_COLORS.keys())
combined_df = pd.read_csv('combined_df.csv')
combined_df
```

```python
combined_df.drop(columns=['Data Url'])
```

```python
#Average Weights for combined + each cut
print(f"Average Weight for all cuts: {np.average(combined_df['Weight'])} carats")

print("Average Weights for each cut:\n")
cuts = ["CUSHION","EMERALD", "HEART", "OVAL","RADIANT","ROUND"]
resultingAvg = []

for cut in cuts:
  resultingAvg.append(np.average(combined_df[combined_df['Shape'] == cut]['Weight']))
  print(f"The average weight for {cut} is {np.average(combined_df[combined_df['Shape'] == cut]['Weight'])} carats")
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
plt.bar(cuts, resultingAvg, color=colors)
plt.xlabel("Cut Shape")
plt.ylabel("Average Weight in carats")
plt.title("Average Weight per cut shape")
plt.show()
```

```python
print(f"std for weight of all cuts: {np.std(combined_df['Weight'])}")

print("std for weight of each cut:\n")
cuts = ["CUSHION","EMERALD", "HEART", "OVAL","RADIANT","ROUND"]
resultingPrice = []

for cut in cuts:
  resultingPrice.append(np.std(combined_df[combined_df['Shape'] == cut]['Weight']))
  print(f"The std for weight of {cut} is {np.std(combined_df[combined_df['Shape'] == cut]['Weight'])}")
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
plt.bar(cuts, resultingPrice, color=colors)
plt.xlabel("Cut Shape")
plt.ylabel("std")
plt.title("std of the weight of each shape")
plt.show()
```

```python
#Max Value
print(f"Highest weight of all cuts: {np.max(combined_df['Weight'])}")

print("std for weight of each cut:\n")
cuts = ["CUSHION","EMERALD", "HEART", "OVAL","RADIANT","ROUND"]
resultingMax = []

for cut in cuts:
  resultingMax.append(np.max(combined_df[combined_df['Shape'] == cut]['Weight']))
  print(f"The heaviest weight of {cut} in carat is {np.max(combined_df[combined_df['Shape'] == cut]['Weight'])}")
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
plt.scatter(cuts, resultingMax, color=colors)
plt.xlabel("Cut Shape")
plt.ylabel("Heaviest Weight in carat")
plt.title("Heaviest Weight of each shape")
plt.show()
```
